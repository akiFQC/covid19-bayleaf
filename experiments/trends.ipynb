```python
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import datetime
```

```python
df = pd.read_csv("../input_data/area_lai_5km_mean.csv", index_col = 0)
df["date"] =pd.to_datetime(df["date"])
df=df.set_index("date")
```

```python
X = df.sort_index(ascending=True).iloc[-30:, 1:].T

```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_norm = ss.fit_transform(X.diff()[1:])
```

```python
print(X_norm.shape)
```

```python
from sklearn.manifold import TSNE

X2d = TSNE(n_components=2).fit_transform(X_norm)
```

```python
dates1 = df.index[(df.index >= datetime.datetime(2019, 2, 1)) &(df.index <= datetime.datetime(2019, 6, 1))]
print(dates1)

dates2 = df.index[(df.index >= datetime.datetime(2020, 2, 1)) &(df.index <= datetime.datetime(2020, 6, 1))]
print(dates2)
```

```python
X1 = df.loc[dates1].sort_index(ascending=True).T
X2 = df.loc[dates2].sort_index(ascending=True).T
```

```python
for  X, ped in zip([X1, X2], ["20190201-0601", "20200201-0601"]):
    ss = StandardScaler()
    X_norm = ss.fit_transform(X.diff()[1:])

    X2d = TSNE(n_components=2).fit_transform(X_norm)
    plt.scatter(X2d[:,0], X2d[:,1], marker=".", s=0.8)
    filename = "../images/"+"tSNE_" + ped+".png"
    plt.savefig(filename)
    plt.show()
```
